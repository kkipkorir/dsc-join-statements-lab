## Select the names of all employees in Boston 

Hint: join the employees and offices tables. Select the first and last name.

```python
# Your code here
q = """
SELECT firstName,lastName,officeCode 
FROM employees
JOIN offices
    USING(officeCode)
WHERE city = 'Boston';
"""

pd.read_sql(q,conn)
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by. Select the office code, city, and number of employees.

```python
# Your code here
q = """
SELECT officeCode, city, COUNT(e.employeeNumber) AS num_employees
FROM offices o
LEFT JOIN employees e
    USING (officeCode)
GROUP BY city
HAVING num_employees = 0
"""

pd.read_sql(q, conn)
```

```python
"""
How many customers are there per employee?
"""

# Your code here
q = """
SELECT e.firstName,e.lastName,COUNT(*) AS num_customers 
FROM employees AS e
JOIN customers AS c
   ON e.employeeNumber = c.salesRepEmployeeNumber
GROUP BY e.firstName,e.lastName;
"""

pd.read_sql(q,conn)
```

```python
"""
How many customers are there per office?
"""

# Your code here
q = """
SELECT o.officeCode, o.city, COUNT(c.customerNumber) AS num_customers
FROM customers AS c
JOIN employees AS e
    ON c.salesRepEmployeeNumber = e.employeeNumber
JOIN offices AS o
    ON e.officeCode = o.officeCode
GROUP BY o.officeCode, o.city
ORDER BY num_customers DESC;
"""

pd.read_sql(q, conn)
```

## Level Up 1: Display the names of every individual product that each employee has sold

Hint: You will need to use multiple `JOIN` clauses to connect all the way from employee names to product names.

```python
# Your code here
q = """
SELECT e.firstName,e.lastName,p.productName
FROM employees AS e
JOIN customers AS c
   ON e.employeeNumber = c.salesRepEmployeeNumber
JOIN orders AS o
    USING(customerNumber)
JOIN orderdetails AS od
    USING(orderNumber)
JOIN products as p
    USING(productCode)
GROUP BY e.firstName, e.lastName

"""

pd.read_sql(q, conn)
```

## Level Up 2: Display the number of products each employee has sold

Alphabetize the results by employee last name.

Hint: Use the `quantityOrdered` column from `orderDetails`. Also, think about how to group the data when some employees might have the same first or last name.

```python
# Your code here
q = """
SELECT e.firstName,e.lastName,COUNT(*) AS num_orders 
FROM employees AS e
JOIN customers AS c
   ON e.employeeNumber = c.salesRepEmployeeNumber
JOIN orders AS o
    USING(customerNumber)
JOIN orderdetails AS od
    USING(orderNumber)
GROUP BY e.firstName, e.lastName
ORDER BY num_orders DESC;
"""

pd.read_sql(q, conn)
```

## Level Up 3: Display the names employees who have sold more than 200 different products

Hint: this is different from the previous question because the quantity sold doesn't matter, only the number of different products

```python
# Your code here
q = """
SELECT e.firstName,e.lastName,COUNT(*) AS num_orders 
FROM employees AS e
JOIN customers AS c
   ON e.employeeNumber = c.salesRepEmployeeNumber
JOIN orders AS o
    USING(customerNumber)
JOIN orderdetails AS od
    USING(orderNumber)
GROUP BY e.firstName, e.lastName
HAVING num_orders > 200
ORDER BY num_orders DESC;
"""

pd.read_sql(q, conn)
```
